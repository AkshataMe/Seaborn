# point plot

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.set_theme(style='darkgrid')
sns.pointplot(data=tips,x='time',y='total_bill',hue='smoker',dodge=True)
```

# Heatmap Plot

```python
flights= sns.load_dataset('flights')
flights.head()
```

```python
plt.figure(figsize=(15,6))
flights=flights.pivot("month",'year','passengers')
sns.heatmap(data=flights,annot=True,fmt='d',linewidths=3.5)
```

# Strip Plot

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(15,6))
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
plt.figure(figsize=(15,6))
sns.stripplot(data=tips,y='day',x='total_bill')
# jitter is use for displacement
```

```python
plt.figure(figsize=(15,6))
sns.stripplot(data=tips,x='day',y='total_bill',linewidth=1.5,size=7.5,hue='sex')
```
